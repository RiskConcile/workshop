<h3>Making a word cloud</h3><br>
From the text below, a <b>word cloud </b>can be made 

```python
text_snippet="European stocks held steady as U.S. equity futures pointed \
to a weaker open and Asian benchmarks declined. \
Oil dropped on signs of oversupply, heading for a seventh weekly loss.\
Gains in banks were counterbalanced by declines in miners, leaving the Stoxx Europe 600\
Index little changed, while shares on the Dow, S&P and Nasdaq all pointed lower with markets\
set to reopen following the Thanksgiving holiday. Chinese Chinese equities led regional declines \
in Asia, with the technology sector weak on concern the U.S. is ratcheting up a campaign against\
Huawei Technologies Co. The dollar and Treasuries were steady while the pound edged lower\
as the Brexit deal heads to the U.K.â€™s Parliament."
```

```python
blob = TextBlob(text_snippet)
word_list = []
for p in blob.tags:
    if p[1]=='NN':
        word_list.append(p[0])
```

```python
print('The first 5 words found:',word_list[0:5])
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
stopwords = set(STOPWORDS)
```

```python
wordcloud = WordCloud(background_color='white',
                      stopwords=stopwords,
                      max_words=30,
                      max_font_size=40,
                      random_state=42).generate(str(text_snippet))
```

```python
fig = plt.figure(1,figsize=(10,10))
plt.imshow(wordcloud)
plt.axis('off');
```
