```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('../../../Data/CoCo.csv')
CoCo =  data['yield'].as_matrix()
```

```python
fig= plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.hist(CoCo,bins=np.arange(4,12,1),histtype='bar',align='mid')
plt.title('bar',fontsize=18)
plt.subplot(2,1,2)
plt.title('step',fontsize=18)
plt.hist(CoCo,bins=np.arange(4,12,1),histtype='step',align='mid')
```

```python
fig= plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.hist(CoCo,bins=np.arange(4,12,1),histtype='step',align='mid')
plt.scatter(CoCo,0*CoCo,marker='x')
plt.grid()
```

```python
fig= plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.hist(CoCo,bins=np.arange(4,12,1),histtype='bar',align='mid')
plt.subplot(2,1,2)
plt.hist(CoCo,bins=20,histtype='bar',align='mid')
```

```python
from sklearn.neighbors import KernelDensity

CoCo = CoCo[:, np.newaxis]
X_range = CoCo
```

```python
fig, ax = plt.subplots(1, 1)
kde = KernelDensity(kernel='tophat', bandwidth=1).fit(CoCo)
log_dens = kde.score_samples(X_range)
ax.fill(X_range[:, 0], np.exp(log_dens), fc='#AAAAFF')
ax.text(-3.5, 0.31, "Tophat Kernel Density")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.neighbors import KernelDensity


#----------------------------------------------------------------------
# Plot the progression of histograms to kernels
np.random.seed(1)
N = 20
X = np.concatenate((np.random.normal(0, 1, 0.3 * N),
                    np.random.normal(5, 1, 0.7 * N)))[:, np.newaxis]
X_plot = np.linspace(-5, 10, 1000)[:, np.newaxis]
bins = np.linspace(-5, 10, 10)

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)
fig.subplots_adjust(hspace=0.05, wspace=0.05)

# histogram 1
ax[0, 0].hist(X[:, 0], bins=bins, fc='#AAAAFF', normed=True)
ax[0, 0].text(-3.5, 0.31, "Histogram")

# histogram 2
ax[0, 1].hist(X[:, 0], bins=bins + 0.75, fc='#AAAAFF', normed=True)
ax[0, 1].text(-3.5, 0.31, "Histogram, bins shifted")

# tophat KDE
kde = KernelDensity(kernel='tophat', bandwidth=0.75).fit(X)
log_dens = kde.score_samples(X_plot)
ax[1, 0].fill(X_plot[:, 0], np.exp(log_dens), fc='#AAAAFF')
ax[1, 0].text(-3.5, 0.31, "Tophat Kernel Density")
```
